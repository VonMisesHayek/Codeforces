```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
user='friedrich_hayek'
all_pages=pd.read_csv(user+'.csv',parse_dates=['When'])
accepted=all_pages[all_pages['Verdict']=='Accepted']
length=len(accepted)
accepted.index=range(0,length)
del accepted['Unnamed: 0']

last=accepted['When'][0]
first=accepted['When'][length-1]
delta=(last-first).days+1
solutions_daily=[]
accum=[]

for i in range(0,delta):
    solutions_daily.append(0)
    accum.append(0)
    
for i in range(0,length):
    solutions_daily[(accepted['When'][length-1-i]-first).days]+=1

suma=0
for i in range(0,delta):
    suma+=solutions_daily[i]
    accum[i]=suma
    
plt.plot(range(0,delta),accum)
plt.xlabel('Days')
plt.ylabel('Accepted Solutions')
plt.title('Problems Solved')
plt.show()
```
